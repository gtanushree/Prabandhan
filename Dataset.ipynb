```python
timeline_ranges = {
    "Last to Last Decade": (datetime(2000, 1, 1), datetime(2010, 12, 31)),
    "Last Decade": (datetime(2011, 1, 1), datetime(2023, 12, 31)),
    "Ongoing": (datetime(2023, 1, 1), datetime(2025, 12, 31))
}

def get_project_duration_days(project_type):
    if any(keyword in project_type.lower() for keyword in ["road", "housing", "infra", "pipeline", "construction"]):
        return random.randint(365, 900)  # 1 to 2.5 years
    elif any(keyword in project_type.lower() for keyword in ["digital", "portal", "dashboard", "licensing"]):
        return random.randint(120, 360)  # 4 months to 1 year
    else:
        return random.randint(180, 720)  # default 6 months to 2 years
    
def get_project_status(start_date, end_date):
    today = datetime.today().date()
    if end_date < today:
        return "Completed"
    elif start_date <= today <= end_date:
        return "Ongoing"
    else:
        return "Upcoming"
    
def get_random_location():
    return random.choice(agra_locations)

def get_project_inventory(project_type):
    inventory_map = {
        "Road Construction": ["Excavators", "Asphalt Mixers", "Labor Crew", "Survey Kits"],
        "Smart Classrooms": ["Projectors", "Tablets", "Wi-Fi Equipment", "Digital Content Licenses"],
        "PHC Expansion": ["Hospital Beds", "Medical Instruments", "Water Supply Tanks", "Solar Lighting"],
        "Water Pipeline": ["PVC Pipes", "Excavation Units", "Leak Testing Kits", "Water Pumps"],
        "Affordable Housing": ["Concrete Mixers", "Bricks", "Steel Rods", "Architect Services"],
        "Waste Management System": ["Garbage Trucks", "Smart Bins", "GPS Devices", "RFID Tags"],
        "eShram Registry": ["Kiosks", "Biometric Scanners", "Data Entry Terminals"],
        "Village MIS": ["Desktops", "Connectivity Units", "Dashboards"],
        "Solar Rooftops": ["Solar Panels", "Inverters", "Mounting Structures", "Cable Harnesses"],
        "Skill Center": ["Laptops", "Tool Kits", "Projectors", "Lab Equipment"]
    }
    resources = inventory_map.get(project_type, ["Generic Tools", "Basic Workforce"])
    return ", ".join(random.sample(resources, k=random.randint(2, len(resources))))
```

```python
data = []

for i in range(10000):
    dept, (proj_type, timeline_tag) = random.choice(list(project_definitions.items()))
    start_range, end_range = timeline_ranges[timeline_tag]

    start_date = start_range + timedelta(days=random.randint(0, (end_range - start_range).days - 180))
    duration = get_project_duration_days(proj_type)
    end_date = start_date + timedelta(days=duration)

    status = get_project_status(start_date.date(), end_date.date())

    location = get_random_location()

    collaborators = collaboration_map.get((dept, proj_type), [])
    other_departments = ", ".join(collaborators) if collaborators else "None"

    inventory = get_project_inventory(proj_type)

    data.append({
        "Project ID": f"PRJ{i+1:05}",
        "Department": dept,
        "Project Type": proj_type,
        "Timeline Tag": timeline_tag,
        "Start Date": start_date.date(),
        "End Date": end_date.date(),
        "Status": status,
        "Location": location,
        "Other Departments Involved": other_departments,
        "Inventory of Available Resources": inventory
    })
```

```python
df = pd.DataFrame(data)
df.to_excel("Dataset.xlsx", index=False)
print("Generated mimic dataset.")
```
